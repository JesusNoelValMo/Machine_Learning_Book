```python
#THIS NOTEBOOK IS FOR PRACTICING CHAPER 2 PF THE BOOK
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import os 
import urllib 
import tarfile

Data_path = os.path.join("datasets", "suicide")
print(Data_path)
```

```python
def Load_Data(Data_path=Data_path):
    csv_path = os.path.join(Data_path, "Suicide.csv")
    return pd.read_csv(csv_path)
    
sui_data = Load_Data()
print(sui_data)
sui_data.describe()
```
