```
import requests 
from urllib.request import urlopen
from lxml import etree
import matplotlib.pyplot as plt
import numpy as np
```

```
# get html from site and write to local file
url = "https://monsterhunter.fandom.com/es/wiki/Nargacuga"
headers = {'Content-Type': 'text/html',}
response = requests.get(url, headers=headers)
raw_html = response.text
html = etree.HTML(raw_html)
```

```
html
```

```
xpath_0 = '//figure[starts-with(@class, "pi-item")]/a/img/@src'
```

```
img_links = html.xpath(xpath_0)
```

```
from PIL import Image
import requests
from io import BytesIO

response = requests.get(img_links[2])
img = Image.open(BytesIO(response.content))
```

```
plt.imshow(img)
```
